# Loading the data

```python
import os
import pandas as pd
import numpy as np
```

```python
def text_to_csv(file_path):
    
    if not os.path.exists(file_path):
        print(f"File {file_path}, doesn't exit!")
        return
    else:
        with open(file_path,'r') as file:
            rows = file.readlines()
        
        rows = map(lambda entry_str: entry_str.strip().split(','), rows)
        
        # create the dataFrame
        df = pd.DataFrame(rows,columns=['feature1','feature2','feature3','feature4','class'])
        df[['feature1','feature2','feature3','feature4']] = df[['feature1','feature2','feature3','feature4']].astype(float)
        df[['class']] = df[['class']].astype(int)
        
        out_path = 'raw-dataset'
        if not os.path.exists(out_path):
            os.makedirs(out_path)
        
        out_file = 'data.csv'    
        df.to_csv(os.path.join(out_path,out_file),index=False)
        return df
df = text_to_csv('./raw-dataset/data_banknote_authentication.txt')
df.head()
```

# Normalize the features

```python
# Load the data
import pandas as pd
df = pd.read_csv('./raw-dataset/data.csv')
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(df.drop('class',axis=1))
norm_data = scaler.transform(df.drop('class',axis=1))
norm_data
```

```python
norm_df = pd.DataFrame(norm_data,columns=['feature1','feature2','feature3','feature4'])
norm_df['class'] = df['class']
norm_df.to_csv('./clean-dataset/normalized.csv',index=False)
norm_df
```
