# Training a basic Model

```python
import pandas as pd


df = pd.read_csv('./clean-dataset/normalized.csv')
df.shape

X = df.drop('class',axis=1).values
Y = df[['class']].values
model = Model(input_shape=X.shape,
              neurones=[4,6,4,1],
              activations=['relu','relu','relu','sigmoid']
              )
losses, acc = model.fit(X,Y,
          batch_size=28,
          nb_iterations=200,
          learning_rate=1e-1,
          val_rate=0.2)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.lineplot(x= range(len(losses)), y=list(map(lambda loss: loss[0],losses)),label= "Train Loss").set(title='Loss evolution', xlabel='Iteration', ylabel='Loss')
sns.lineplot(x= range(len(losses)), y=list(map(lambda loss: loss[1],losses)),label= "Validation Loss")
```

```python
sns.lineplot(x= range(len(losses)), y=list(map(lambda loss: loss[0],acc)),label= "Train accuracy").set(title='Accuracy evolution', xlabel='Iteration', ylabel='Accuracy')
sns.lineplot(x= range(len(losses)), y=list(map(lambda loss: loss[1],acc)),label= "Validation accuracy")
```
